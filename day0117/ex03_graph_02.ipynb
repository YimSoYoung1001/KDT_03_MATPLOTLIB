# [데이터 시각화 맛보기]
- x,y축 레이블 지정하기
- x,y축 눈금 설정하기
- 한글 폰트 기설정하

(1) 모듈 로딩

```python
import pandas as pd
import matplotlib.pyplot as plt             # 그래프 관련 모듈
import matplotlib as mpl                    # 패키지 관련
import matplotlib.font_manager as fm        # 패키지 리소스 즉 자원 중 폰트 관련 모듈
```

(2) 데이터 준비

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = range(1, 100, 10)                        #1~99 
y = [value ** 2 for value in x]       #리스트 컴프리헨션
```

(3) 그래프 그리기

- [기본] 라인/선 그래프
    * 시간의 변화를 표현해주는 그래프
    * 연속적/순서 있는 데이터를 시각화할 경우 주로 사용됨

```python
#plot(data) => y축 데이터로 사용
#plot(data1, data2) => x축에 data1, y축에 data2 데이터로 사용
#반환값 : 라인/선 그래프 객체 리스트에 담겨서 반환됨
ret = plt.plot(x, y)
type(ret), len(ret), type(ret[0])               #그래프는 리스트에 담겨서 준다

# 파이썬에서 표준 출력 기능 내장 함수 => print() 동일 역할
plt.show()
```

```python
ret[0].get_xdata()
```

- 라인/선 그래프에서 형식 즉 포맷 지정 => str 타입의 포맷 설정
    * '색상형태선스타일' ==> 'ro--' : red circle dot-line
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b - blue, k - black
    * 대표적으로 많이 사용되는 형태 : o - 동그라미 , ^ - 위로 삼각형, < - 왼쪽 삼각형, > - 오른쪽 삼각형, v - 아래로 삼각형
    * 대표적으로 많이 사용되는 선스타일 : - 실선, -- 점선, -.- 점이 포함된 

```python
ret = plt.plot(x, y, 'rv-.')
```

```python
ret = plt.plot(x, y, 'ro--')
```

```python
ret = plt.plot(x, y, 'r^--')
```

```python
ret = plt.plot(x, y, 'g^--')
```

```python
x2 = [d//2 for d in x]
y2 = [d//2 for d in y]
```

    * 그래프에 제목 설정

### X, Y 축 레이블 설정 및 눈금 설정
- x, y 축 데이터에 대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태 설정 => plt.xticks(), plt.yticks(), plt.tick_param()

```python
# 그래프 그리기
plotList = plt.plot(x, y, 'ro--', x2, y2, 'g^')
plt.title("[뷁]")

# x, y 출 레이블
plt.xlabel("값", loc = 'right', fontsize = 'xx-large', color = 'purple', rotation = 120)
plt.ylabel("결과", loc = 'top', fontsize = 'xx-large', color = 'gray', rotation = 1500)

# x, y 축 눈금 설정
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'red')
plt.yticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'blue')
```

### 범례 설정
- 여러 개의 그래프를 표시하는 경우 그래프별 식별을 위해서 사용

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^')
#plotList = plt.plot(y, 'ro--', y2, 'g^', label = '그냥 결과')
plt.title("[뷁]")

# x, y 출 레이블
plt.xlabel("값", loc = 'right', fontsize = 'xx-large', color = 'purple', rotation = 120)
plt.ylabel("결과", loc = 'top', fontsize = 'xx-large', color = 'gray', rotation = 1500)

# x, y 축 눈금 설정
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'red')
plt.yticks(color = 'blue')

# 범례 설정
plt.legend(['red 결과', 'green 결과'], loc = 'center')    #첫줄에 그래프그리기에서 라벨이 설정되어있다면 자동으로 추가됨. 그게아니면 명시적으론 넣어야
```

### 텍스트 추가
- 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text(시작x, 시작y,  출력테스트)

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^')
#plotList = plt.plot(y, 'ro--', y2, 'g^', label = '그냥 결과')
plt.title("[뷁]")

# x, y 출 레이블
plt.xlabel("값", loc = 'right', fontsize = 'xx-large', color = 'purple', rotation = 120)
plt.ylabel("결과", loc = 'top', fontsize = 'xx-large', color = 'gray', rotation = 1500)

# x, y 축 눈금 설정
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E'], color = 'red')
plt.yticks(color = 'blue')

# 범례 설정
plt.legend(['red 결과', 'green 결과'], loc = 'center')    #첫줄에 그래프그리기에서 라벨이 설정되어있다면 자동으로 추가됨. 그게아니면 명시적으론 넣어야

# 그래프 안에 텍스트 출력
plt.text(2, 2000, 'text_add', backgroundcolor = 'yellow', color = 'red')

box = {'facecolor':'.75', 'edgecolor':'k', 'boxstyle':'round', 'alpha':0.3}
plt.text(2, 4000, 'text_add', bbox=box)
```
