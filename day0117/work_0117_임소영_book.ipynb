# work 01.17. 임소영 책 과제 (3장 격자까지)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
```

```python
plt.title('A polymonomial')
plt.plot(X, Y, c = 'k')
plt.show()
```

```python
# 라텍스 사용하기
plt.title('$f(x) = \\frac{1}{4}(x+4)(x+1)(x-2)$')
plt.plot(X, Y, c = 'k')
plt.show()
```

```python
# 각축에 레이블 추가
plt.title('Power curve for airfoil KV873')

plt.xlabel('Air Speed')
plt.ylabel('Total drag')

plt.plot(X, Y, c = 'k')
plt.show()
```

```python
# 텍스트 추가
plt.text(-0.5, -0.25, 'Brackmard minimum')
plt.plot(X, Y, c = 'k')
plt.show()
```

```python
# 텍스트 추가 + 경계상자 제어
box = {'facecolor':'.75', 'edgecolor':'k', 'boxstyle':'round'}
plt.text(-0.5, -0.25, 'Brackmard minimum', bbox = box)
plt.plot(X, Y, c = 'k')
plt.show()
```

```python
# 화살표 추가
plt.annotate('Brackmard minimum', 
             ha = 'center', va = 'bottom', 
             xytext = (-1.5, 3.),
             xy = (0.75, -2.7),
             arrowprops={'facecolor':'black', 'shrink':0.05})
plt.plot(X, Y)
plt.show()
```

```python
# 범례 추가
X = np.linspace(0, 6, 1024)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.xlabel('X')
plt.ylabel('Y')

plt.plot(X, Y1, c = 'k', lw = 3., label = 'sin(X)')
plt.plot(X, Y2, c = '.5', lw = 3., ls = '--', label = 'cos(X)')

plt.legend()
plt.show()
```

```python
# 격자 추가
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.plot(X, Y, c = 'k')
plt.grid(True, lw = 2, ls = '--', c = '.75')   #격자추가
plt.show()
```
